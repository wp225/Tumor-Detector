```python
from Tumor_Detector.recognizer.src.model import resnet
from Tumor_Detector.recognizer.src.utils.data_loader import DataLoader
```

```python
!unzip '/content/drive/MyDrive/Django_full.zip'
```

```python
train_path = '/content/Django_full/archive/Training'
test_path = '/content/Django_full/archive/Testing'
loader = DataLoader(train_path,test_path)
train_df, test_df = loader.load_data()
print(f'Train Classes: {train_df.class_names}')
print(f'Test Classes: {train_df.class_names}')
```

```python
input_shape = (224,224,1)
n_classes = 4
model = resnet(input_shape,n_classes)
model.summary()
```

```python
import mlflow

ngrok_forwarding_url = "6.tcp.eu.ngrok.io:14041"

postgres_uri = f"postgresql+psycopg2://postgres:24680@{ngrok_forwarding_url}/mlflow"

mlflow.set_tracking_uri(postgres_uri)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

mlflow.tensorflow.autolog()

with mlflow.start_run(run_name='epoch-10-mymodel'):
  model.fit(
      train_df,
      validation_data = test_df,
      epochs = 20,
      batch_size = 32
  )

```

```python
model.save('/content/drive/MyDrive/mymodel/mymodel.keras')
```
